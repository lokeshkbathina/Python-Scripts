Face detection is a trivial problem for humans to solve and has been solved reasonably well by classical feature-based techniques, such as the cascade classifier.<br> More recently, DL methods have achieved SOTA results on standard face detection datasets. One example is the **Multi-task Cascade Convolutional Neural Network** (MTCNN).

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from mtcnn.mtcnn import MTCNN
```

```python
pic = mpimg.imread('/content/photo.jpg')
print(pic.shape)
plt.imshow(pic)
```

loading the model/detector (using default weights)

```python
model = MTCNN()
```

```python
face = model.detect_faces(pic)
```

```python
face
```

```python
x, y, width, height = face[0]['box']
```

```python
fig, ax = plt.subplots()
ax.imshow(pic)
# creating a rectangle patch (bounding box) for the detected face
rect = patches.Rectangle((x, y), width, height, fill=False, color='red', linewidth = 1.5)
ax.add_patch(rect)
plt.show()
```
